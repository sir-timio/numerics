```python
from math import sqrt, log
import matplotlib.pyplot as plt
import numpy as np
from numpy import sqrt, sum, power
```

```python
xs = np.array([1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5])
ys = np.array([3.21, 2.95, 4.06, 4.03, 5.39, 5.97, 6.51, 6.77, 7.79])
```

```python
def arith_mean(a, b):
    return (a + b) / 2

def sqrt_mean(a, b):
    return sqrt(a*b)

def harmonic_mean(a, b):
    return (2 / (1/a + 1/b))
```

```python
def compute_sqrt_error(ys, pred_ys, normalize=False):
    e = sum(power(ys-pred_ys, 2))
    if normalize:
        return e / len(ys)
    return e
```

```python
plt.plot(xs, ys, '.')
```

```python
xa = arith_mean(xs[0], xs[-1])
xg = sqrt_mean(xs[0], xs[-1])
xh = harmonic_mean(xs[0], xs[-1])
```

```python
ya = arith_mean(ys[0], ys[-1])
yg = sqrt_mean(ys[0], ys[-1])
yh = harmonic_mean(ys[0], ys[-1])
```

```python
A = sum([x**2 for x in xs])
B = sum(xs)
C = sum([x*y for x, y in zip(xs, ys)])
D = sum(xs)
E = len(xs)+1
F = sum(ys)
```

```python
a = round((B*F - C*E) / (B*D - A*E), 2)
b = round((C*D - A*F) / (B*D - A*E), 2)
```

```python
def f(x, a=a, b=b):
    return a*x+b
```
